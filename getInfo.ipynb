# 3. Clean the JSON files to a CSV table

```python
import json
import os
import pandas as pd
import numpy as np
import re
```

```python
def extract_info(json_file_name):
    f = open('../data/scholors_duke/complete' + json_file_name, 'r')
    j = json.load(f)
    d = dict(j)
    
    # duid -- Duke Unique ID
    duid = json_file_name.split('.')[0]
    
    # uri -- URI
    if d['uri'] == '':
        uri = ''
    else:
        uri = d['uri']
    
    # label -- Current Label
    if d['label'] == '':
        label = ''
    else:
        label = d['label']
    
    # school -- School Name
    if len(d['positions']) == 0:
        school = ''
    else:
        school = d['positions'][0]['attributes']['schoolLabel']
    
    # title -- Title
    if d['title'] == '':
        title = ''
    else:
        title = d['title']
    
    # education -- Past Education
    if len(d['educations']) == 0:
        return [duid, uri, label, school, title, '', '', '']
    elif len(d['educations']) == 1:
        education1 = d['educations'][0]['attributes']['institution']
        return [duid, uri, label, school, title, education1, '', '']
    elif len(d['educations']) == 2:
        education1 = d['educations'][0]['attributes']['institution']
        education2 = d['educations'][1]['attributes']['institution']
        return [duid, uri, label, school, title, education1, education2, '']
    else:
        education1 = d['educations'][0]['attributes']['institution']
        education2 = d['educations'][1]['attributes']['institution']
        education3 = d['educations'][2]['attributes']['institution']
        return [duid, uri, label, school, title, education1, education2, education3]
```

```python
def main():
    df = pd.DataFrame(columns=['duid', 'uri', 'label', 'school', 'title', 'education1', 'education2', 'education3'])
    json_file_names = os.listdir('./scholors_info/')
    for json_file_name in json_file_names:
        if re.match(r"[0-9]+(.json)", json_file_name):
            print('============={}============='.format(json_file_name))
            info = extract_info(json_file_name)
            df.loc[len(df)] = info
        else:
            continue
    
```

```python
main()
```
