# 2. Use the DUIDs to crawl JSON via Scholars widgets API

```python
import requests
import json
import pandas as pd
```

## Load DUIDs

```python
df = pd.read_csv('./data/duid.csv')
duid_df = df[pd.notnull(df['id'])]
duid_df
```

## Crawl JSON data

Data Mode
- Complete
- Publication

```python
def crawl(duid, name, data_mode):
    url = 'https://scholars.duke.edu/widgets/api/v0.9/people/{}/5.json?uri={}'.format(data_mode, duid)
    headers = {"Accepct": "application/json"}
    response = requests.get(url=url, headers=headers)
    if response.status_code == 404:
        return (name, duid)
    json_info = response.json()
    return json_info

def save_json(data_mode, duid, content):
    dir_str = './data/scholors_duke/{}/{}.json'.format(data_mode, duid)
    with open(dir_str, 'w') as outfile:
        json.dump(content, outfile)
```

```python
def main(duid_df, data_mode):
    counter = 0
    for i in range(duid_df.shape[0]):
        duid = duid_df.loc[i, 'id']
        duid  = str(int(duid)).zfill(7)
        name = duid_df.loc[i, 'name']
        
        no_info_list = list()
        
        json_content = crawl(duid, name, data_mode)
            
        if type(json_content) == tuple:
            no_info_list.append(json_content)
        else:
            counter += 1
            save_json(data_mode, duid, json_content)
        
        if i % 200 == 0:
            print('=============={}completed =============='.format(i))       
    print('=============={}JSON files containing {} infomation in total crawled =============='.format(counter, data_mode))
```

```python
main(duid_df, 'complete')
```

```python
main(duid_df, 'publications')
```
